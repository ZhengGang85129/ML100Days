```python
import os
import numpy as np
import pandas as pd
from pandas import DataFrame
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
```

### 如果沒有想法，可以先嘗試找出剛剛例子中提到的問題的答案
#### 資料的 row 數以及 column 數

```python
number_of_row = app_train.shape[0]
number_of_col = app_train.shape[1]
print("Number of rows is : %s" % (number_of_row)) 
print("Number of columns is : %s" % (number_of_col)) 
```

#### 列出所有欄位

```python
dataFrame = app_train.head(number_of_row)
dataFrame
```

#### 截取部分資料

```python
row_start=90
row_end=95
#col用兩種方式去取索引位置
#(1)用 loc(label)
print(app_train['AMT_INCOME_TOTAL'])
col_start='AMT_INCOME_TOTAL'
col_end='BASEMENTAREA_MEDI'
app_train.loc[row_start:row_end,col_start:col_end]
```

```python
#2 用iloc
columns = app_train.columns
col_start=columns.get_loc('AMT_INCOME_TOTAL')
col_end=columns.get_loc('BASEMENTAREA_MEDI')
app_train.iloc[row_start:row_end+1,col_start:col_end+1]
```
